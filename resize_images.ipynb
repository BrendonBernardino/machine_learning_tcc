```python
from PIL import Image
import os
```

```python
input_folder = "/content/drive/MyDrive/Dataset_TCC/ESP_Slave"
output_folder = "/content/drive/MyDrive/Dataset_TCC/ESP_Slave/images_288"

os.makedirs(output_folder, exist_ok=True)

new_size = (288, 288)

for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)

        try:
            img = Image.open(input_path)

            # Redimensionar mantendo a proporção (se necessário, pode mudar para `.resize(new_size)`)
            img = img.resize(new_size, Image.LANCZOS)

            img.save(output_path)
            print(f"✅ Redimensionado: {filename} -> {output_path}")

        except Exception as e:
            print(f"❌ Erro ao processar {filename}: {e}")

print("\n✅ Todas as imagens foram redimensionadas!")
```
